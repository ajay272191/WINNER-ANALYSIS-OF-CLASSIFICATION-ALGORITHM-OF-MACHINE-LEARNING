```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
xgData=pd.read_csv('xgboostAnalysis.csv')
xgData.head()
```

```python
adData=pd.read_csv('adaboostAnalysis.csv')
adData.head()
```

```python
randomData=pd.read_csv('randomAnalysis.csv')
randomData.head()
```

```python
final_dataSet = pd.concat([xgData,adData,randomData], axis=1)
final_dataSet.head()
```

```python
final_dataSet.drop(['dataSet_ada','dataSet_random'],axis=1,inplace=True)
#final_dataSet.head()
```

```python
 final_dataSet.rename(columns={"dataSet_xg": "dataSet"},inplace=True)
 final_dataSet.head()
```

# Winner

```python
#mean value random boost classifier
final_dataSet['accuracy_random'].mean()
```

```python
#mean value for Adaboost classifier
final_dataSet['accuracy_ada'].mean()
```

```python
#mean value for xgBoost classifier
final_dataSet['accuracy_xg'].mean()
```

# winner :: Random forest 

# winner analysis for individual dataset: 

```python
#random forest accuracy distribution
fig=sns.kdeplot(final_dataSet['accuracy_random'],shade=True)

#xgboost accuracy distribution
fig=sns.kdeplot(final_dataSet['accuracy_xg'],shade=True)

# adaboost accuracy distribution
fig=sns.kdeplot(final_dataSet['accuracy_ada'],shade=True)
sns.set(rc={'figure.figsize':(25,10)},font_scale=2)
```

```python
#random forest accuracy distribution
sns.barplot(x='dataSet',y='accuracy_random',data=final_dataSet)
sns.set(rc={'figure.figsize':(25,10)})
```

```python
# adaboost accuracy distribution
sns.barplot(x='dataSet',y='accuracy_ada',data=final_dataSet)
sns.set(rc={'figure.figsize':(25,10)})
```
