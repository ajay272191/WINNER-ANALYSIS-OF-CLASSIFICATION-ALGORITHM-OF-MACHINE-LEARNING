# Cotegory wise Analysis of winner

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
xgData=pd.read_csv('xgboostAnalysis.csv')
xgData.head()
```

```python
randomData=pd.read_csv('randomAnalysis.csv')
randomData.head()
```

```python
adData=pd.read_csv('adaboostAnalysis.csv')
adData.head()
```

```python
final_dataSet = pd.concat([xgData,adData,randomData], axis=1)
final_dataSet.head()
```

```python
final_dataSet.drop(['dataSet_ada','dataSet_random'],axis=1,inplace=True)
#final_dataSet.head()
```

```python
final_dataSet.rename(columns={"dataSet_xg": "dataSet"},inplace=True)
final_dataSet.head()
```

# Analysis on Bio Informative

```python
bio_data=final_dataSet.iloc[[0, 1, 2, 3, 7, 8, 9, 10, 18]]
bio_data
```

```python
#random forest accuracy distribution
fig=sns.kdeplot(bio_data['accuracy_random'],shade=True)

#xgboost accuracy distribution
fig=sns.kdeplot(bio_data['accuracy_xg'],shade=True)

# adaboost accuracy distribution
fig=sns.kdeplot(bio_data['accuracy_ada'],shade=True)

fig.set(title='Accuracy Distribution')
sns.set(rc={'figure.figsize':(25,10)},font_scale=2)
```

```python
#mean value random boost classifier
random_mean=bio_data['accuracy_random'].mean()
random_mean
```

```python
ada_mean=bio_data['accuracy_ada'].mean()
ada_mean
```

```python
xg_mean=bio_data['accuracy_xg'].mean()
xg_mean
```

# winner for Bio informative dataset : Random Forest Classifier
